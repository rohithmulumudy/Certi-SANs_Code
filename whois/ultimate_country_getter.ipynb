```python
import pickle
nodes = [a.replace("\n", "") for a in open("clf_test/ninety.txt", "r").readlines()]
data1 = pickle.load(open("ninety_features.pickle", "rb"))
data2 = pickle.load(open("ninety_tld_features.pickle", "rb"))
```

```python
from MySQLdb import _mysql
db = _mysql.connect(user="root", passwd="Shar_mistha1", db="places")
def get_country(name):
    blk1 = data1.get(name, {})
    blk2 = data2.get(".".join(name.split(".")[-2:]), {})

    if blk1 == {} and blk2 == {}:
        return "Unknown"

    for k in blk1.keys():
        if "country" in k.lower():
            return blk1[k]
        elif "state" in k.lower() or "province" in k.lower():
            try:
                db.query('SELECT countries.code FROM regions LEFT JOIN countries ON regions.country_id = countries.id WHERE LOWER(regions.name) = "{}";'.format(blk1[k].lower().strip()))
                r = db.store_result().fetch_row()
                if r != tuple():
#                     print(r[0][0], 1)
                    return r[0][0].decode().upper()
            except:
                pass
        elif "city" in k.lower():
            try:
                db.query('SELECT countries.code FROM cities LEFT JOIN countries ON cities.country_id = countries.id WHERE LOWER(cities.name) = "{}";'.format(blk1[k].lower().strip()))
                r = db.store_result().fetch_row()
                if r != tuple():
#                     print(r[0][0], 2)
                    return r[0][0].decode().upper()
            except:
                pass
            
    for k in blk2.keys():
        if "country" in k.lower():
            return blk2[k]
        elif "state" in k.lower() or "province" in k.lower():
            try:
                db.query('SELECT countries.code FROM regions LEFT JOIN countries ON regions.country_id = countries.id WHERE LOWER(regions.name) = "{}";'.format(blk2[k].lower().strip()))
                r = db.store_result().fetch_row()
                if r != tuple():
#                     print(r[0][0], 3)
                    return r[0][0].decode().upper()
            except:
                pass
        elif "city" in k.lower():
            try:
                db.query('SELECT countries.code FROM cities LEFT JOIN countries ON cities.country_id = countries.id WHERE LOWER(cities.name) = "{}";'.format(blk2[k].lower().strip()))
                r = db.store_result().fetch_row()
                if r != tuple():
#                     print(r[0][0], 4)
                    return r[0][0].decode().upper()
            except:
                pass
        
    return "Unknown"
            
        
    
```

```python
import tqdm
countries = {}
for n in tqdm.tqdm(nodes):
    countries[n] = get_country(n)
```
