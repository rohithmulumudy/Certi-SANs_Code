```python
# Load data

import pandas as pd

df = pd.read_csv("ca_san_count.csv")
df
```

```python
labels = df["CA"]
counts = df.drop(columns=["CA"])

counts
```

```python
list(labels)
```

```python
cnt_np = counts.to_numpy()
```

```python
# K-Means Clustering

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=30).fit(cnt_np)

clusters = {
    i:[] for i in range(30)
}

for i in range(len(list(labels))):
    clusters[list(kmeans.labels_)[i]].append(list(labels)[i])
```

```python
clusters
```

```python
import tqdm

def make_clusters(k, data):
    kmeans = KMeans(n_clusters=k).fit(data)
    return kmeans.inertia_

inertias = [make_clusters(k, cnt_np) for k in tqdm.tqdm(range(2, 150))]
```

```python
# Find the elbow

import matplotlib.pyplot as plt

plt.plot(list(range(2, 150)), inertias)
plt.show()
```

```python
# Take a closer look
plt.plot(list(range(2, 20)), inertias[:18])
plt.show()
```

```python
# So elbow is at 6

kmeans = KMeans(n_clusters=6).fit(cnt_np)
clusters = {
    i:[] for i in range(6)
}

for i in range(len(list(labels))):
    clusters[list(kmeans.labels_)[i]].append(list(labels)[i])
```

```python
print([len(c) for c in clusters.values()])
```

```python
print(clusters)
```

```python
import statistics

total_counts = dict(zip(list(labels), list(counts.sum(axis=1))))
for idx, cluster in clusters.items():
    totals = [total_counts[a] for a in cluster]
    try:
        std = statistics.stdev(totals)
    except:
        std = 0
    print("Cluster: {:2}\tSize: {:4}\tMean: {:.6}\tStd Dev: {:.6}\tMedian: {:.6}".format(idx, len(cluster),
         float(statistics.mean(totals)),
         float(std),
         float(statistics.median(totals))))
```
